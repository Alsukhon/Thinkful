```python
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
# %matplotlib inline
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')
```

For this challenge I will be using Breast Cancer Wisconsin (Prognostic) Data Set:

```python
data = pd.read_csv('data.csv')
data.head()
```

Let's check for missing values:

```python
data.isnull().sum()
```

```python
# drop unnecessary columns
data = data.drop(columns=['id','Unnamed: 32'])
data.head()
```

```python
# split data to train & test
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(50, 4, ),max_iter=10000,tol=0,n_iter_no_change=10000) 
mlp.fit(X_train, y_train) 
print('Score is: {}', mlp.score(X_test, y_test)) 
print('Crossval score is: {}', cross_val_score(mlp, X_test, y_test, cv=10))
```

So the MLP has only reached 96% after convergence. Let's see how Random Forests performs:

```python
rfc = RandomForestClassifier(n_estimators=100)
model = rfc.fit(X_train, y_train)
print('Score is: {}', model.score(X_test, y_test))
print('Crossval score is: {}', cross_val_score(model, X_test, y_test, cv=10))
```

We can see that MLP has performed better than Random Forests. However, Random Forests had a stable score across all CV partitions, going as low as 85%, where MLP did go down to 65% on several CV partitions.
