Let us take a look at the correlation of our data, moving the target column to the front and ordering rows by their correlation with it:

```python
# make the correlation matrix.
corr_mat_houses = df.corr()

# get a list of columns
cols = list(corr_mat_houses)

# move the column to head of list using index, pop and insert
cols.insert(0, cols.pop(cols.index('saleprice')))

# use ix to reorder
corr_mat_houses = corr_mat_houses.ix[:, cols]

corr_mat_houses.sort_values(by=['saleprice'],ascending=False)
```

The columns we have chosen as features before can be seen in the top rows. None of the negatively correlated features are significant enough to include within the model's features. To visualize this table, let's draw a correlation matrix, filtering out correlation less than a set threshold for a focused view on relevant features:

```python
def draw_heatmap(df,corr_thresh=.7):
    
    # make the correlation matrix.
    corr_mat = df.corr()

    # move the column to head of list using index, pop and insert
    cols = list(corr_mat)
    cols.insert(0, cols.pop(cols.index('saleprice')))

    # reorder
    corr_mat = corr_mat.ix[:, cols]
    corr_mat.sort_values(by=['saleprice'],ascending=False)

    # figure aesthetics
    plt.figure(figsize=(18,15))
    plt.plot()
    
    # draw the heatmap using seaborn.
    sns.heatmap(corr_mat[(corr_mat > corr_thresh)], square=True, annot=True, linewidths=.5)
    plt.title("Correlation Matrix (Houses)")
    printmd("**Correlation Matrix (filtered for correlation above {}):**".format(corr_thresh))
    plt.show()
    
    # return count of 
    features = corr_mat[(corr_mat > corr_thresh)]
    features = features.index[features['saleprice'] >= corr_thresh].tolist()
    features.remove('saleprice')
    print('features are: {}'.format(features))
    return features

corr_thresh = 0.6
features = draw_heatmap(df,corr_thresh)


    
# del corr_mat_houses
```

An initial set of features is produced based on a correlation threshold. In the heatmap we can see that some of these features are correlated, so we'd like to point them out and remove ones that do not add much information to the model:

```python
corr_mat_houses.loc['lotarea','saleprice']
corr_mat_houses.index[corr_mat_houses['saleprice'] >= 0.7].tolist()
```

```python
# Produce a list of {column, index} of cells with 0.7 or higher correlation
corr_pairs = []
for col in corr_mat_houses.columns:
    corr_feats = corr_mat_houses.index[corr_mat_houses['saleprice'] >= 0.7].tolist()

# Drop sets with no values that are >= 0.7 (added as {column})
corr_pairs = [i for i in corr_pairs if len(i)>1]

# Features that correlate 0.7 or more with the target
corr_w_target = [i for i in corr_pairs if 'saleprice' in i]

# Features that correlate 0.7 or more with each other
corr_features = [i for i in corr_pairs if 'saleprice' not in i]

# Drop duplicates
corr_features = [i for i in corr_features if ({i[0],i[1]} not in corr_features[:corr_features.index(i)])]
```
