```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r"Mall_Customers.csv")
```

```python
print(df.columns)
df.head()
```

```python
# Drop the useless CustomerID column
df = df.drop(columns=["CustomerID"])

# Shorten column names
df = df.rename(columns={"Annual Income (k$)":"income","Spending Score (1-100)":"spending"})
```

```python
# Plot the distribution of Annual Income (in thousand $)

# sns.set(style=)
```

```python

sns.set(style="darkgrid")
sns.distplot(df["income"])
plt.show()
```

```python
sns.kdeplot(df["income"],shade=True)
plt.show()
```

```python
plt.plot(df["income"])
plt.show()
```

```python
sns.boxplot(df["income"])
plt.show()
```

```python
sns.relplot(x="income",y="spending",data=df)
plt.show()
```

```python
sns.relplot(x="income",y="spending",hue="Gender",data=df)
plt.show()
```

```python
sns.relplot(x="spending",y="income",kind="line",data=df)
plt.show()
```

```python
sns.jointplot(x="spending",y="Age",data=df)#,kind="reg")
```
