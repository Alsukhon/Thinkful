```python
stfd = pd.read_csv(r"avtemp.csv")
```

```python
stfd.head()
stfd = stfd.rename(columns={"Ann":"annual_avg"})
```

```python
mich1 = stfd[stfd["City"] == "DETROIT"]
mich2 = stfd[stfd["City"] == "FLINT"]
mich3 = stfd[stfd["City"] == "LANSING"]

mich = pd.concat([mich1,mich2])
mich = pd.concat([mich,mich3])
```

```python

fig, axs = plt.subplots(1, 2, figsize=(9, 3))
sns.catplot(x='City',y='annual_avg', kind='bar',orient='v', data=mich,ax=axs[0])#bar(names, values)
sns.relplot(x='Nov',y='City', size="Nov",sizes=(25,250), data=mich, ax=axs[1])#scatter(names, values)
fig.suptitle('Temperatures of Major Michigan Cities')


# axs[0].set_title("Annual Temperature Avg Dist")

# axs[1].set_title("Annual Temperature")
# plt.tight_layout()
plt.show()
```

```python
g = sns.PairGrid(mich)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
