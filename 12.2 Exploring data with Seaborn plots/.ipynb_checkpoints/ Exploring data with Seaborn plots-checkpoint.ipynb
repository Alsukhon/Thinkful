```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"Mall_Customers.csv")
```

```python
stfd = pd.read_csv(r"avtemp.csv")
```

```python
stfd.head()
stfd = stfd.rename(columns={"Ann":"annual_avg"})
```

```python
mich1 = [stfd = stfd["City"] == "DETROIT"]
mich2 = [stfd["City"] == "FLINT"]
mich3 = [stfd["City"] == "LANSING"]

mich = pd.concat(mich1,mich2)
mich = pd.concat(mich,mich3)
```

```python
fig, axs = plt.subplots(ncols=3)

sns.catplot(x='annual_avg',y='City', kind='bar',orient='h', data=mich,ax=axs[0])
# axs[0].set_title("Annual Temperature Avg Dist")
sns.relplot(x='Nov',y='City', size="Nov",sizes=(25,250), data=mich, ax=axs[1])
# axs[1].set_title("Annual Temperature")
# plt.tight_layout()
# plt.show()
```

```python
g = sns.PairGrid(mich)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend();
```
