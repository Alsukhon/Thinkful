```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

#create an empty dataframe and fill it with 50 integers between 1 and 100
df = pd.DataFrame()
df["random_nums"] = np.random.randint(1,100,50)
```

```python
#plot our column
plt.plot(df["random_nums"])
plt.ylabel("Frequency")
plt.xlabel("Random Integers")
plt.title("First Plot")
plt.show()
```
