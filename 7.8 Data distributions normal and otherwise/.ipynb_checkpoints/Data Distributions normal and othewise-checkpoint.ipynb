### For each distribution:

Generate a random variable with 100 datapoints.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#normal
normal_dist = np.random.normal(5,2, 100)

#binomial
binomial_dist = np.random.binomial(23,.45, 100)

#poisson
poisson_dist = np.random.normal(5, 100)

#F
f_dist = np.random.normal(5,45, 100)

#exponential
exponential_dist = np.random.normal(0.01, 100)

#pareto
pareto_dist = np.random.normal(4, 100)

distDict = {"normal_dist":normal_dist,"binomial_dist":binomial_dist,"poisson_dist":poisson_dist,"f_dist":f_dist,"exponential_dist": exponential_dist,"pareto_dist":pareto_dist}
```

Graph the variable using a histogram.

```python
for i in distDict:
    print(i)
    plt.hist(distDict[i])
    plt.show()
```

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

```python
for i in distDict:
    mean = np.mean(distDict[i])
    std_dev = np.std(distDict[i])
    print(i)
    plt.hist(distDict[i])
    # Add a vertical line at the mean.
    plt.axvline(mean, color='r', linestyle='solid', linewidth=2)

    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(mean + std_dev, color='b', linestyle='dashed', linewidth=2)

    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(mean - std_dev, color='g', linestyle='dashed', linewidth=2) 

    # Print the histogram.
    plt.show()
```

Evaluate whether the descriptive statistics provided useful information about the variable. Can you identify any common characteristics of the distributions that could be usefully described using the mean and/or standard deviation, versus the ones that could not?

The normal, binomial, and f distributions share the common characteristic in that their values are mostly centered around the mean, along with standard deviation being symmetrical around the mean.

As for possion, exponential, and pareto distributions, the mean and standard deviation are clustered in one point. Which provides no value for statistical inference about the data.

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1. 

```python
normal_1 = np.random.normal(5,0.5, 100)
normal_2 = np.random.normal(10,1, 100)
```

Add them together to create a third variable.

```python
normal_all = normal_1 + normal_2
```

Graph the third variable using a histogram.

```python
plt.hist(normal_all)
plt.show()
```

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

```python
    mean = np.mean(normal_all)
    std_dev = np.std(normal_all)
    
    plt.hist(normal_all)
    # Add a vertical line at the mean.
    plt.axvline(mean, color='r', linestyle='solid', linewidth=2)

    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(mean + std_dev, color='b', linestyle='dashed', linewidth=2)

    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(mean - std_dev, color='g', linestyle='dashed', linewidth=2) 

    # Print the histogram.
    plt.show()
```

Evaluate the descriptive statistics against the data.

Even when two datasets with a different mean and standard deviation each are added together, their normal distribution is still inherited, albeit slightly varies, into the third dataset made from merging both of them.

The mean and standard deviations still describe the center of the data and how datapoints vary around that center.
