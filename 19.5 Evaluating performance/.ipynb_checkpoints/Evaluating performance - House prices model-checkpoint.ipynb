# Model 1: The original model

```python
df = df[['SalePrice','OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF','FullBath','YearBuilt','YearRemodAdd']]

features = list(df.columns)
features.remove('SalePrice')
print('features (len: {}): {}\ntarget: SalePrice'.format(len(features),{i for i in features}))

X = df[features]

Y = df['SalePrice']

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

# Model 2: Add renovation period

This model will consider the period between year built and renovation year:

```python
features = list(df.columns)
features.remove('SalePrice')

X = df[features]


# try to match the correlation numbers 
# to the scatterplots above to get a 
# sense of what each magnitude of 
# correlation can look like.

# Remove YearRemodAdd and YearBuilt, and add the difference between them
X['renovPeriod'] = df['YearRemodAdd'] - df['YearBuilt']
X = X[['OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF','FullBath','renovPeriod']]
print('features (len: {}): {}\ntarget: SalePrice'.format(len(X.columns),{i for i in X}))


Y = df['SalePrice']

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

By removing year built and year renovated and adding the difference in between, R2, Adjusted R-squared, and the F-statistic are lower, while AIC and BIC are higher. This model is not better than the first one. Still, we may can do some more feature engineering for this model. Let's take a look at its confusion matrix:

```python
# create a function that does feature engineering: custom function
#***********************************************************************
# make the correlation matrix.
corr_mat_houses = df.corr()
corr_mat_houses = corr_mat_houses#[corr_mat_houses > .7]


plt.figure(figsize=(10,8))

plt.plot(1,1)
# draw the heatmap using seaborn.
sns.heatmap(corr_mat_houses, square=True, annot=True, linewidths=.5)
plt.title("Confusion matrix (Houses)")

plt.show()
```

# Model 3: combining two features

We see that GarageCars and GarageArea are strongly correlated and therefore one of them should be removed. Let's try to combine them and see whether that can help our model:

```python
# add a new column, area_p_car
df['area_p_car'] = df['GarageArea']/df['GarageCars']

# convert nan/inf to numerical
df['area_p_car'] = df['area_p_car'].apply(np.nan_to_num)



# df = df[['SalePrice','OverallQual','GrLivArea','GarageCars','GarageArea','TotalBsmtSF','FullBath','YearBuilt','YearRemodAdd']]

features = list(df.columns)
features.remove('SalePrice')
features.remove('GarageCars')
features.remove('GarageArea')
print('features (len: {}): {}\ntarget: SalePrice'.format(len(features),{i for i in features}))


X = df[features]

Y = df['SalePrice']

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

# Model 4: one feature less than the original Model 1

That did not help the model. We will remove area_p_car and put back GarageCars because it is a -very small- bit more correlated with the target:

```python
df.drop(columns=["area_p_car", "GarageArea"],inplace=True)
```

```python
features = list(df.columns)
features.remove('SalePrice')
print('features (len: {}): {}\ntarget: SalePrice'.format(len(features),{i for i in features}))

X = df[features]

Y = df['SalePrice']

X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()

results.summary()
```

With one less feature we were able to get back to the best evaluation metrics; those of the original model. Moreover, the F-statistic is higher, meaning it is superior to the original model.
