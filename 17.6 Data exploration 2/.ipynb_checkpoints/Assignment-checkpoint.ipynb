1. Are there any differences between the genders, ethnicities, and parental level of education with respect to their performances in exams?


```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
# draw the heatmap using seaborn.
sns.barplot(data["gender"], data["math score"])
plt.title("Math score per gender")

plt.subplot(1,3,2)
# draw the heatmap using seaborn.
sns.barplot(data["gender"], data["reading score"])
plt.title("Reading score per gender")

plt.subplot(1,3,3)
# draw the heatmap using seaborn.
sns.barplot(data["gender"], data["writing score"])
plt.title("Writing score per gender")

plt.show()
```

2. Are there any differences between the lunch types with respect to their performances in exams? If there are, how do you explain this?

```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
# draw the heatmap using seaborn.
sns.barplot(data["lunch"], data["math score"])
plt.title("Math score per lunch")

plt.subplot(1,3,2)
# draw the heatmap using seaborn.
sns.barplot(data["lunch"], data["reading score"])
plt.title("Reading score per lunch")

plt.subplot(1,3,3)
# draw the heatmap using seaborn.
sns.barplot(data["lunch"], data["writing score"])
plt.title("Writing score per lunch")

plt.show()
```

It seems that those who have standard lunch perform better on exams. Lunch type itself may not have a direct relationship with grades, because it may reveal something about the student's living situation, which could be hard enough to affect studying conditions. But again, students who have lunch of less quality may have their perforamnce affected due to lack of nutritios lunch content.

3. Does the test preparation course seem to have an effect on the exam performances?


```python
plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
# draw the heatmap using seaborn.
sns.barplot(data["test preparation course"], data["math score"])
plt.title("Math score per test preparation course")

plt.subplot(1,3,2)
# draw the heatmap using seaborn.
sns.barplot(data["test preparation course"], data["reading score"])
plt.title("Reading score per test preparation course")

plt.subplot(1,3,3)
# draw the heatmap using seaborn.
sns.barplot(data["test preparation course"], data["writing score"])
plt.title("Writing score per test preparation course")

plt.show()
```

4. Which 2 exam scores are most correlated with each other?


```python
print('math with: \n reading:')
print (data["math score"].corr(data["reading score"]))
print ('\n writing:')
print (data["math score"].corr(data["writing score"]))

print('reading with : \n math:')
print (data["reading score"].corr(data["math score"]))
print ('\n writing:')
print (data["reading score"].corr(data["writing score"]))

print('writing with: \n reading:')
print (data["writing score"].corr(data["reading score"]))
print ('\n math:')
print (data["writing score"].corr(data["math score"]))
```

Reading and writing correlate the most with each other.
